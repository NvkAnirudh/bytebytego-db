```python
from llama_index.readers.web import RssReader

reader = RssReader()

rss_feed = "https://blog.bytebytego.com/feed"
documents = reader.load_data([rss_feed])
```

```python
docs = {}

for doc in documents:
    if doc.id_ not in docs:
        docs[doc.id_] = {}
    
    docs[doc.id_]['title'] = doc.metadata['title']
    docs[doc.id_]['source'] = doc.metadata['link']
    docs[doc.id_]['published_date'] = doc.metadata['date']
    docs[doc.id_]['text'] = doc.text_resource.text
```
